````python
['A','B','C','A','B','C']

````

<h3>Intersection of Sets</h3> 


Find the intersection of set <code>A</code> and <code>B</code>.


````python
A={1,2,3,4,5}
B={1,3,9, 12}
````

<details><summary>Click here for the solution</summary>

```Python
A & B

```
OR

```Python
A.intersection(B)
```

</details>
